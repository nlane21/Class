```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#Load Data
url = 'https://data.cityofnewyork.us/api/views/uip8-fykc/rows.csv?accessType=DOWNLOAD'

df = pd.read_csv(url)

#Filter to only include K and B
#df_filtered = df[(df['ARREST_BORO'] == 'K') | (df['ARREST_BORO'] == 'B')]
df_filtered = df[(df['ARREST_BORO'].isin(['K','B']))]
df_filteredtwo = df[(df['ARREST_BORO'].isin(['M','K']))]

#Plot
plt.figure(figsize=(7,6))
sns.countplot(x='ARREST_BORO', data=df_filtered,palette='muted')
plt.title('Number of Arrests by Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Arrests')
plt.show()


```

```python
#Display plot map for K and M Arrest Boro
sns.countplot(x='ARREST_BORO', data=df_filteredtwo,palette='bright')
plt.title('Number of Arrests by Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Arrests')
plt.show()
```
