```python
import pandas as pd

# Load the data
url = 'https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv'
df = pd.read_csv(url)

# Look at the data
print(df.shape)
print(df.columns)
print(df.head())
```

```python
#Cleanse the data

# Check for missing values
print(df.isnull().sum())

# Check data types
print(df.dtypes)
```

```python
# Remove rows with missing or non-numeric mean scale scores
df['Mean Scale Score'] = pd.to_numeric(df['Mean Scale Score'], errors='coerce')
mean_by_subject = df.groupby('Subject')['Mean Scale Score'].mean().sort_values(ascending=False)
print(mean_by_subject)
```

```python
# Remove rows with missing or non-numeric mean scale scores
# The original column name might have extra spaces or a slightly different name.
# Check the actual column name in the dataframe using df.columns
# Assuming the column name is 'Mean Scale Score' but with leading/trailing spaces:
df['Mean Scale Score'] = pd.to_numeric(df['Mean Scale Score'].str.strip(), errors='coerce')
# If the column name is different, replace 'Mean Scale Score' with the actual name. For example:
# df['Mean Score'] = pd.to_numeric(df['Mean Score'].str.strip(), errors='coerce')

mean_by_subject = df.groupby('Subject')['Mean Scale Score'].mean().sort_values(ascending=False)
print(mean_by_subject)
```

```python
# Remove rows with missing or non-numeric mean scale scores
# The original column name might have extra spaces or a slightly different name.
# Check the actual column name in the dataframe using df.columns
# Assuming the column name is 'Mean Scale Score' but with leading/trailing spaces:
# df['Mean Scale Score'] = pd.to_numeric(df['Mean Scale Score'].str.strip(), errors='coerce')
# If the column name is different, replace 'Mean Scale Score' with the actual name. For example:

# Check the actual column name and replace with the correct one
# print(df.columns) # Uncomment to see the column names
# Assuming the actual column name is 'Mean Score':
df['Mean Score'] = pd.to_numeric(df['Mean Score'].astype(str).str.strip(), errors='coerce')

mean_by_subject = df.groupby('Subject')['Mean Score'].mean().sort_values(ascending=False) # Use the correct column name
print(mean_by_subject)
```

```python
# Remove rows with missing or non-numeric mean scale scores
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors='coerce')
df.dropna(subset=['Mean Score'])
```

```python
# Average score by year
mean_by_year = df.groupby('School Year')['Mean Scale Score'].mean()
print(mean_by_year)
```

```python
# Rename the column
df = df.rename(columns={'Year': 'School Year'})

# Average score by year
mean_by_year = df.groupby('School Year')['Mean Score'].mean()
print(mean_by_year)
```

```python
# Group by school and average their scores
top_schools_averages = df.groupby('School Name')['Mean Score'].mean().sort_values(ascending=False).head(10)
print(top_schools_averages)
```

```python
top_schools_averages_subject = df.groupby(['School Name', 'Regents Exam'])['Mean Score'].mean().sort_values(ascending=False).head(10)
print(top_schools_averages_subject)
```

```python
highest_average = df.groupby('School Name')['Mean Score'].mean().idxmax()
print(highest_average)
```

```python
school = df[df['School Name'] == highest_average]
print(school)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group and calculate mean score by subject
mean_by_subject = df.groupby('Regents Exam')['Mean Score'].mean().sort_values()


# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_by_subject.values, y=mean_by_subject.index, palette='viridis')
plt.title('Average Mean Scale Score by Subject')
plt.xlabel('Mean Score')
plt.ylabel('Subject')
plt.tight_layout()
plt.show()
```
